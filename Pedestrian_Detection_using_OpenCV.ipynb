```python
import cv2
import imutils
from google.colab.patches import cv2_imshow
```

```python
# Initializing the HOG person detector
hog = cv2.HOGDescriptor()
hog.setSVMDetector(cv2.HOGDescriptor_getDefaultPeopleDetector())
```

```python
# Reading the image
image = cv2.imread('./drive/MyDrive/complex-C2.jpg')
```

```python
# Resize the image
image = imutils.resize(image, width=min(400, image.shape[1]))
```

```python
# Convert the image to grayscale for better processing
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

```python
# Applying histogram equalization to enhance contrast
gray = cv2.equalizeHist(gray)
```

```python
# Detecting all the regions in the image that has pedestrians inside it
(regions, _) = hog.detectMultiScale(image, winStride=(4,4), padding=(4,4), scale=1.05)
```

```python
# Drawing the regions in the image
for (x,y,w,h) in regions:
    cv2.rectangle(image,(x,y), (x+w, y+h), (0,0,255),2)
```

```python
# Showing the output image
cv2_imshow(image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
